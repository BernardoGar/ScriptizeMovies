```python
def austenize(sentences, words):
    finalset=[]
    for sentence in sentences:
        lx=[]
        mini=0
        flag=1
        for word in words:
            if sentence.find(" "+word+" ")<mini:
                flag=0
                break
            else:
                if mini>0:
                    lx.append(sentence[mini:sentence.find(" "+word+" ")])
                mini=sentence.find(word)+len(" "+word)
        if flag==1:
            finalset.append(lx)
    return finalset

def austenizenearest(sentences, wordsor,model, topn):
    finalset=[]
    wordsmatrix=[]
    for word in wordsor:
        wordsvector=[]
        model_word_vector = np.array((model[word]), dtype='f')
        most_similar_words = model.most_similar( [ model_word_vector ], [], topn)
        wordsmatrix.append([i[0] for i in most_similar_words])
    intento=0
    print(len(wordsor))
    while intento<(topn**len(wordsor)):
        f=intento
        words=[]
        for vec in wordsmatrix:
            words.append(vec[int(f%topn)])
            f=(f-f%topn)/topn
        ns=austenize(sentences, words)
        for s in ns:
            finalset.append(s)
        intento=intento+1
    return finalset

def extractMrMrs(sentence):
    newcharacters=[]
    charactermarks=['mr ', 'mrs ', 'sir ', 'ms ', 'miss ']
    for char in charactermarks:
        stt=0
        while sentence.find(char, stt)>=0:
            stt=sentence.find(char, stt)
            end=sentence.find(" ", sentence.find(char, stt)+len(char)+1)
            nchar=sentence[stt:end]
            newcharacters.append(nchar)
            stt=stt+1
    return list(set(newcharacters))


def listwords(sentence):
    liston=sentence.split(' ')
    charactermarks=['mr', 'mrs', 'sir', 'ms', 'miss']
    s=0
    while s<len(liston):
        if liston[s] in charactermarks and s+1<len(liston):
            liston[s]=liston[s]+" "+liston[s+1]
            liston.pop(s+1)
        s=s+1
    return liston
```
